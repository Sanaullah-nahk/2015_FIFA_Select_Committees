Reading the csv file into a pandas dataframe, taking the first column as a list of the names of the FIFA members, popping the first value as its a zero, then creating a graph with these names.

```python
FIFA_graph.remove_edges_from(nx.selfloop_edges(FIFA_graph))
print(len(FIFA_graph.nodes()))
print(len(FIFA_graph.edges()))
```

Plotting the first visualization of the graph. Not expecting much, mostly used for a comparison later on.

```python
plt.figure(figsize=(10,10))
nx.draw(FIFA_graph, node_color="black",edge_color="grey", node_size=40, width=0.25)
plt.show()
```

Following statistics confirm the above graph, mostly that most members degrees are around the mode and average, with a few outliers.

Following two blocks are the first test of Force Atlas for this graph, only used for a comparison later in this section as the image doesnt tell us too much.

```python
# setup ForceAtlas2 for FIFA Graph
# code taken in part from ForceAtlas2 github but adjusted for our needs
forceatlas2 = ForceAtlas2(
                          # Behavior alternatives
                          outboundAttractionDistribution=True,  
                          edgeWeightInfluence=1,
                          # Performance
                          jitterTolerance=0.2, # Tolerance
                          barnesHutOptimize=True,
                          barnesHutTheta=0.1,
                          # Tuning
                          scalingRatio=0.2,
                          strongGravityMode=True,
                          gravity=1,
                          # Log
                          verbose=True) 

# Calculate positions with the algorithm
positions = forceatlas2.forceatlas2_networkx_layout(FIFA_graph, pos=None, iterations=2000) 
```

```python
f = plt.figure()
f.set_size_inches(10, 10)
nx.draw_networkx_edges(FIFA_graph, positions, alpha=0.2)
nx.draw_networkx_nodes(FIFA_graph, positions, alpha=1, node_size=10)
plt.axis('off')
plt.title("ForceAtlas2 FIFA Network")
plt.show()
```

## Downloading Wikipedia pages in order to align FIFA member names with Wikipedia article names - Ben

The following code was the initial downloading of the Wikipedia pages in order to inform the fixing of the names in the CSV file to correspond with the persons name on Wikipedia. This had to occur as the Wikipedia API demands perfect spelling of pages to return them. Further work on the Wikipedia pages used in the Community and Sentiment Analysis sections are shown in the Community Analysis section below.

```python
wikilist = open("fixed-wikilist.txt", "r")
data = wikilist.read()
list_data = data.split("\n")
```

```python
for i in range(len(list_data)):
  list_data[i] = list_data[i].lower()
```

```python
for i in range(len(list_data)):
  list_data[i] = list_data[i].title()
```

```python
import json
def buildWikiQuery(page):
    base = "http://en.wikipedia.org/w/api.php/?"
    action = "action=query"
    title = "titles=%s" % page 
    content = "prop=revisions&rvprop=content"
    format = "format=json"
    query = "%s%s&%s&%s&%s" % (base, action, title, content, format)
    query = query.replace(" ","_")
    return query
```

```python
def fetchContent(query):
    result = 'No content'
    revision_key = 'revisions'
    try:
        response = urllib.request.urlopen(query)
    except UnicodeEncodeError:
        return 0
    try:
      json_reponse = json.load(response)
      pages = json_reponse['query']['pages']
      page_key = list(pages.keys())[0]
      page_content = pages[page_key]
      if revision_key in page_content:
          content = pages[page_key][revision_key]
          result = content[0]['*'].encode('utf8')
      return result
    except json.JSONDecodeError:
      return 0
```

```python
def saveToFile(name, content):
    if content == 'No content':
        print('Empty page for:', name)
    else:
        try:
          with open(r'%s' % (name),'wb') as f:
              f.write(content)
              f.close()
        except OSError:
          print('Empty page for:', name) 
```

```python
def fetch():
    wikipages = list_data
    for i, page_name in enumerate(wikipages):
        wiki_query = buildWikiQuery(page_name)
        if fetchContent(wiki_query) != 0:
          wiki_content = fetchContent(wiki_query)
        else:
          continue
        saveToFile(page_name, wiki_content)
```

```python
# Function to download Wiki text as text file all files
import json
import os
from urllib.parse import quote_plus
import urllib
def DownloadPages(title):
    
    title_new = "titles=" + quote_plus(title)
    baseurl = "https://en.wikipedia.org/w/api.php?"
    #content = "prop=revisions&rvprop=content"
    content = "prop=extracts&exlimit=1&explaintext"  #for more simple plan text (just wii text)
    action = "action=query"
    dataformat = "format=json"
    query = "%s%s&%s&%s&%s" % (baseurl,action,title_new,content,dataformat)
    query = query.replace(" ","_") # DONE till here 
    response = urllib.request.urlopen(query)
    source = response.read()
    wikitext = source.decode('utf-8')
    json_character = json.loads(source)
    
    #for text file
    with open("text_files_2/"+title+'.txt', 'w+') as fh:
        fh.write(wikitext)
    #for Json file  just in case also save it as Json file
    p = "json files/" + title.replace(' ', '_') + '.json' 
    with open(p, 'w') as json_file:
        
        json.dump(json_character, json_file)
```

# Eextract *`(Read .txt)`*,  Transform *`(Clean wiki text)`*,  Load *`(save clean text to its respective community list)`*. *- Sana*
> Eextract all text releated to a specisific community, clean the text and save it its corresponding list for further processing. this we would have 10 lists (one list of each community) and and each list contains all the "Cleaned" wiki texts of a perticulaer community  

```python
import re
from sklearn.feature_extraction import _stop_words
import nltk
nltk.download('omw-1.4')
nltk.download('wordnet')
nltk.download('punkt')
from nltk.stem import WordNetLemmatizer


lemmatizer = WordNetLemmatizer()
from nltk.stem.porter import PorterStemmer
#Stop_Words= _stop_words.ENGLISH_STOP_WORDS

nltk.download('stopwords')
from nltk.corpus import stopwords
Stop_Words=stopwords.words('english')

custom = ['fifa', 'football'] #custom stop_words
Stop_Words.extend(custom)

# Clean the text
def text_processing(text):
    
    # Remove URL from text
    #text = re.sub(r"(@\[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)|^rt|http.+?", "", text)
    text = re.sub(r'http\S+', '', text)
    
    #Rmove tweeter mentions from text (for later tweets)
    text = re.sub("\S*@\S*\s?","", text)
    
    #Remove any hash tag from text
    text = re.sub("#[A-Za-z0-9_]+","", text)
    
    # Remove inside brackets i.e (Chinese: 傅满洲 Fù Mǎnzhōu) etc.
    text = re.sub("[\(\[].*?[\)\]]", "", text)
    
    # remove punctuation 
    text = "".join([c for c in text 
                    if c not in string.punctuation])
    # lowercase handel
    text = "".join([c.lower() for c in text])
    
    # remove stopwords
    text = " ".join([w for w in text.split() 
                     if w not in Stop_Words])
    
    #remove year or refrence numbers
    text = ''.join([i for i in text if not i.isdigit()])
    
    # lematizing 
    text = " ".join([lemmatizer.lemmatize(w) for w in text.split()])
    
    # Stemming
    #stemmer = PorterStemmer()
    #text = " ".join([stemmer.stem(w1) for w1 in text.split()])
    return text
```

```python
# WIKI TEXT | ETL
# Eextract (Read) Transform(Clean) and load(Save)
import os
count=0
success1=0
no_text_list=[]

community_name_list = set()
for root, dirs, files in os.walk('/content/gdrive/MyDrive/Final Fifa Files/fifa_txt_file/'): #walk in folder
    for file in files:
        try:

            #name = file.removesuffix('.txt') #get file name from folder
            name = file[:len(file) - 4]      # if  python version is below 3.9
            if name in list(communities.keys()): #check if file is part of any community
                number=communities[name] #check which community it belongs to
                with open(os.path.join(root, file), 'r', encoding='utf-8') as f: #read
                    text = f.read()
                    character_data=json.loads(text)

                    dict_vals = character_data['query']['pages'].keys()
                    page_id = list(dict_vals)[0] #ID
                    character_data_content = character_data['query']['pages'][page_id]['extract'] #get main text data

                    character_data_content = text_processing(character_data_content) # CLEAN DATA USING ABOVE FUCNTION

                    command="community"+str(number)+".append(character_data_content)" # Save CLEAN data
                    comand_set="community_name_list"+".add('community"+str(number)+"')" # maintain "community_name_list" for later
                    exec(comand_set) 
                    exec(command)
                    success1+=1
        except:
            count+=1
            no_text_list.append(name) # just incase if i want to see names of no wiki text people.

print(f"No wiki text found for {count} files")
print(f"text extrection succeed for {success1} files")
```

# (3) **`TF-IDF` calculation and  `'Word Count'` representation.** - *Sana*

```python
from sklearn.feature_extraction.text import TfidfVectorizer
def community_TFIDF(community):
    vectorizer = TfidfVectorizer()    # create instance of TF IDF vectoriser class
    vecs = vectorizer.fit_transform(community)    # train and create vectorizer
    feature_names = vectorizer.get_feature_names_out()    # unique words to display it later in DF or ...
    dense = vecs.todense()    # this is out actual TFIDF vectoriser
    lst1 = dense.tolist()    #Same to list
    df = pd.DataFrame(lst1, columns=feature_names)    # DF to visualize 
    
    return dense, df     # return TFIDF vector and DF with all TF-IDF values for each word in each document. 
```

```python
# Functions To Download Tweets
import snscrape.modules.twitter as sntwitter

# 50 Most recent for each member.
def get_tweet(query):
    query = query + " lang:en"
    tweet1=[]
    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
        #tweets.append(tweet.content)
        if len(tweet1)==50:
            break
        else:
            tweet1.append(tweet.content)
    list_to_string = ' '.join(tweet1)
    return list_to_string


# 50 tweets for each member between 2015 - 2017.
def get_tweet_2015(query):
    query = query + " lang:en until:2017-12-30 since:2015-02-01"
    tweet1=[]
    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
        #tweets.append(tweet.content)
        if len(tweet1)==50:
            break
        else:
            tweet1.append(tweet.content)
    list_to_string = ' '.join(tweet1)
    return list_to_string
```

## Analysis of the text

Utilis function for compute the sentiment of a text usong NLTK.

```python
# Function to calculate the sentiment of a single word
def word_sentiment(w,d):

  #S1 needs to be dataset with the name as index
    if not(w.isalpha()) :
        #print('error: non alphab char')
        return
    w = w.lower()
    if w in d.index:
        res = d.loc[w,:].happiness_average
        return float(res)
    else :
        #print('word not found')
        return 


# Finction to calculate the sentiment of a text
def calc_sent(text : str,d):
  
    count = 0 # number of visited words
    tot_sent = 0
    fdist = nltk.FreqDist(text.split(' '))
    for w,t in fdist.items():
        #print(w, count, tot_sent)
        sent = word_sentiment(w,d)
        if sent:
            count += 1*t
            tot_sent += sent*t
    if count == 0 : count = 1
    return tot_sent/count
```

Analysis of the set of wikipedia page, firstly with the original dictionary and later with the custom.

```python
# TRY ONE: normal sentiment analysis

dict_sent = {}
for root, dirs, files in os.walk('/content/gdrive/MyDrive/text files'):
  for file in files: 
    txt_file = open('/content/gdrive/MyDrive/text files' + '/' + str(file))
    sent_res = calc_sent(txt_file.read(),S1)
    if sent_res > 0: dict_sent[file[:-4]] = sent_res


m = np.percentile(list(dict_sent.values()), 50)
p = np.percentile(list(dict_sent.values()), [10,90])
print(f'The mean value is {round(m, 4)}.\nThe 10th percentile is {round(p[0], 4)} and the 90th percentile is {round(p[1], 4)}.')

b = np.linspace(4.7,6.25,25)
plt.hist(dict_sent.values(), bins = b, color= 'orange')
plt.show()
```
